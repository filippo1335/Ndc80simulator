```python
import sys
import numpy as np

polymerLenght = 10
headN = 1


# DEFINE SYSTEM

def polymergenerator(polymerl, headl, mass=1):
    polymer = np.array((range(polymerl), np.zeros(polymerl), np.zeros(polymerl))).T
    # heads = np.array([range(headl), np.zeros(headl), np.zeros(headl)]).T
    heads = np.array((5, 0.15, 0.15))
    """ future functionality
    head_surface = heads.copy()
    head_surface[0] += 1
    centers_of_mass = (head_surface[0] + heads) / 2"""
    return polymer, heads #, centers_of_mass, head_surface


# DEFINE FORCES
def forcefield(a, b, epsilon=0.2, sigma=0.15, mass = 1, random = 0.1):
    dif = a - b
    dist = np.sqrt(np.sum(np.square(dif), axis=1))
    f = 24 * epsilon * (-2*(sigma**12 / dist**13) + (sigma**6 / dist**7))
    vector = dif / np.array((dist, dist, dist)).T
    forcevector = np.sum(vector * np.array((f, f, f)).T, axis=0) +  np.random.uniform(-random, random, (3,))
    # a = forcevector/mass 
    return forcevector


# DEFINE INTEGRATION ALG
def verlet(positions, velocities, accelerations, dt, random):
    positions = positions + velocities * dt + (accelerations/2) * (dt**2)
    accelerationst1 = forcefield(polymerX, positions, random = random) #+ np.random.uniform(-0.02, 0.02, velocities.shape)
    velocities =  velocities + (accelerationst1 + accelerations) * (dt/2)
    return positions, velocities, accelerationst1

#def ang_integrate(center, surf_1, surf_2, ang_vel, ang_acc, dt, offsets, masses):
    
```

```python
polymerX, headX = polymergenerator(polymerLenght, headN)
forcevectors = forcefield(polymerX, headX)
np.set_printoptions(threshold=sys.maxsize)
print(forcevectors)
```

```python
polymerX, headX = polymergenerator(polymerLenght, headN)
np.set_printoptions(threshold=sys.maxsize)

vX = np.array((0,0,0))
aX = np.array((0,0,0))

steps = 200000
print_every = 1

#track history positions, velocities and accelerations 
headhist = np.zeros((int(steps/print_every), 3))
velhist = np.zeros((int(steps/print_every), 3))
acchist = np.zeros((int(steps/print_every), 3))

# RUN
for i in range(steps):
    headX, vX, aX = verlet(headX, vX, aX, 0.01, random = 0.1)
    if i%10000 == 0: print(headX, vX[1], aX[1]) # sanity check output to console
    headhist[i] = headX
    velhist[i] = vX
    # acchist[i] = aX
```

```python
import matplotlib.pyplot as plt
time = np.arange(200000)
fig, ax = plt.subplots(2,gridspec_kw={'height_ratios': [4, 1], "width_ratios": [1]}, figsize=(10,10))
pos = ax[0].scatter(headhist[::20,0], headhist[::20,1], cmap="gist_rainbow", c=time[::20], s=0.1)
ax[0].set_xlim(4.6, 5.4)
ax[0].set_ylim(-0.4, 0.4)
ax[0].set_aspect("equal", adjustable="box")
vel = ax[1].scatter(time[::20], velhist[::20,0], cmap="hsv", c=time[::20], s=0.1)
fig.colorbar(pos, label='timestep')
```

```python
with open("trajectory.pdb", "w") as f:
    for t, positions in enumerate(headhist[::200]):  # Save every 100th frame
        f.write(f"MODEL     {t + 1}\n")
        f.write(f"ATOM  {1:5d}  C   UNK     1    {positions[0]:8.3f}{positions[1]:8.3f}{positions[2]:8.3f}  1.00  0.00           C\n")
        f.write("ENDMDL\n")
```
